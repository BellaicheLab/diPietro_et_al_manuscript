```python
from __future__ import print_function, unicode_literals, absolute_import, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from tifffile import imread
from csbdeep.utils import axes_dict, plot_some, plot_history

from csbdeep.io import load_training_data
from csbdeep.models import Config, CARE
import os
```

```python
BaseDir = '/data/u934/service_imagerie/v_kapoor/CurieTrainingDatasets/MouseClaudia/'
NPZdata = 'Mouse400ClaudiaSegmentModel'+ '.npz'

ModelDir ='/data/u934/service_imagerie/v_kapoor/CurieDeepLearningModels/MouseClaudia/'
ModelName = 'Mouse400ClaudiaSegmentationUNet-53'
load_path = BaseDir + NPZdata 
```

```python

(X,Y), (X_val,Y_val), axes = load_training_data(load_path, validation_split=0.1, verbose=True)

c = axes_dict(axes)['C']
n_channel_in, n_channel_out = X.shape[c], Y.shape[c]
```

```python
plt.figure(figsize=(12,5))
plot_some(X_val[:5],Y_val[:5])
plt.suptitle('5 example validation patches (top row: source, bottom row: target)');
```

```python
config = Config(axes, n_channel_in, n_channel_out, probabilistic = False, unet_n_depth=5,unet_kern_size=3,train_epochs= 100, train_batch_size = 30,  train_reduce_lr={'patience': 5, 'factor': 0.5})
print(config)
vars(config)
```

```python
model = CARE(config = config, name = ModelName, basedir = ModelDir)
model.load_weights(ModelDir + ModelName + '/' + 'weights_best.h5')
```

```python
history = model.train(X,Y, validation_data=(X_val,Y_val))
```

```python
print(sorted(list(history.history.keys())))
plt.figure(figsize=(16,5))
plot_history(history,['loss','val_loss'],['mse','val_mse','mae','val_mae']);
```

```python
plt.figure(figsize=(12,7))
_P = model.keras_model.predict(X_val[25:30])
if config.probabilistic:
    _P = _P[...,:(_P.shape[-1]//2)]
plot_some(X_val[25:30],Y_val[25:30],_P,pmax=99.5)
plt.suptitle('5 example validation patches\n'      
             'top row: input (source),  '          
             'middle row: target (ground truth),  '
             'bottom row: predicted from source');
```
